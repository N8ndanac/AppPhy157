#### PART 1 - Get mean normalized spectrum *(25 pts.)*

Plot the mean spectrum by first normalizing the spectra and plot the mean and standard deviation. For this, you may use the ``normalize`` function in scikit-learn's preprocessing module [(documentation)](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.normalize.html). Describe the mean spectrum and note wavelengths with the highest variation across the sample. What is the reason for this variation?


We first import the necessary modules from sklearn

````python
from sklearn import preprocessing, decomposition
````

Next, we proceed to normalize and plot the spectra against the wavelengths. This is known as the mean spectrum.

````python
#Extract the wavelengths from the data set
wl = data["wavelength"]
#Extract and normalize the spectra from the data set
normalized_data = preprocessing.normalize(data["spec_recons"])

#Compute the mean and standard deviation along the primary axis from the normalized spectra
mean = normalized_data.mean(0)
std = normalized_data.std(0)

#Plot the mean and standard deviation
plt.plot(wl, mean, color="mediumseagreen", label="mean")
plt.fill_between(wl, mean - std, mean + std, color="aquamarine", label="mean $\pm$ std")
#Formatting
plt.xlabel("Wavelength [$\AA$]")
plt.ylabel("Flux")
plt.legend()
plt.title("Mean spectrum")
#Displaying
plt.show()
````

The largest spikes in variation for the mean spectrum are located at around $\lambda = 3750$ Å and around $\lambda = 6500$ Å. There is also a region of high variation around $\lambda = 5000$ Å.

#### PART 2 - Get principal components *(75 pts.)*

Perform Principal Component Analysis (PCA) and visualize the top principal components:
1. Get the first 4 principal components using the ``PCA`` function in scikit-learn's decomposition module [(documentation)](https://scikit-learn.org/0.15/modules/generated/sklearn.decomposition.PCA.html#sklearn.decomposition.PCA). Make sure to use the normalized spectra as input. *(25 pts.)*


````python
#Perform PCA
pca = decomposition.PCA(n_components=4)
#Use fit_transform instead to include dimensionality reduction
transformed_data = pca.fit_transform(normalized_data)
````

Let us review the dimensionality of the original data set,

````python
print(normalized_data.shape)
````

Again, the original data set consists of 4000 spectra represented by 1000 attributes. Another way to imagine this is each spectrum lives in a 1000-dimensional space. Each dimension represents an attribute. We are interested in the trends between these attributes that could help us identify each spectrum.

Let us compare this to the transformed data set,

````python
print(transformed_data.shape)
````

After performing PCA, we have effectively reduced our 1000-dimensional space into a 4-dimensional space. It is still difficult to imagine, but a lot easier to work with.

Each dimension, known as a principal component, represents a trend between some attributes. The first principal component, PC1, accounts for the largest variations in the data set. PC2 accounts for the second largest, and so on.

2. Plot the first two principal components (coefficient 1 in the x axis and coefficient 2 in the y axis). Make sure to label the axes and good to set x and y axis limits to be equal. *(25 pts.)*

````python
#Plotting
plt.scatter(transformed_data[:, 0], transformed_data[:, 1], s=2, color="purple", alpha=0.2)
#Formatting
plt.xlim(-0.7, 1.1)
plt.ylim(-0.7, 1.1)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA projection of the spectra")
#Displaying
plt.show()
````

We have plotted the normalized, transformed data points. Each point corresponds to a point from our original data set. However, their position in this plot represents some combination of the original 1000 attributes. Points closer in this plot share similarities based on the original attributes, whereas more distant points have more different attributes.

However, it still looks messy! Let's compare it to the original data set.

````python
#Plotting
plt.scatter(normalized_data[:, 0], normalized_data[:, 1], s=2, color="goldenrod", alpha=0.2)
#Formatting
plt.xlabel("Flux1")
plt.ylabel("Flux2")
plt.title("First two principal components of the normalized data")
#Displaying
plt.show()
````

We can observe an approximately linear trend from the previous data. Intuitively, this is something that will be simplified in the process of dimensionality reduction. Also take note of the difference in scales. The original normalized data ranges from $-0.03$ to $0.05$, while the transformed data ranges from $-0.7$ to $1.1$. The larger scale makes variations more prominent.

We can also take the dot product of the components divided by its magnitude,

````python
print(pca.components_[0]@pca.components_[1]/np.sqrt(pca.components_[0]@pca.components_[0])/np.sqrt(pca.components_[1]@pca.components_[1]))
````

This number is very close to zero. Hence, we expect the first principal axis to have a slope close to zero, while the second principal axis should have a very small positive slope.

From the plot of our transformed data, these predictions are correct.

3. Plot the first 4 eigenvectors together with the mean spectrum. Add an arbitrary y-offset between the spectra for clarity. *(25 pts.)*

````python
#Plotting
for i in range(4):
  plt.plot(wl, 5*pca.components_[i] + i*2 + 2, label="Eigenspectrum " + str(i + 1))
plt.plot(wl, 20*mean, label="Mean spectrum")
#Formatting
plt.xlabel("Wavelength [$\AA$]")
plt.ylabel("Scaled flux (with offset)")
plt.legend(bbox_to_anchor=(1,1))
plt.title("Mean spectrum and eigenspectra")
#Displaying
plt.show()
````

Each eigenspectrum corresponds to a principal component. We can see that the variations in the mean spectrum correspond to variations in each eigenspectrum.

For example, the aforementioned spike at around $\lambda = 6500 $ Å is largely due to eigenspectrum 3, while the spike at around $\lambda = 3750 $ Å is largely due to eigenspectra 1 and 2. Meanwhile, eigenspectra 1, 2, and 4 all contribute to the spike at around $\lambda = 5000 $ Å. Each of these eigenspectra correspond to certain characteristics of galaxies.

From our data, we can characterize these galaxies based on their principal components and eigenspectra.
