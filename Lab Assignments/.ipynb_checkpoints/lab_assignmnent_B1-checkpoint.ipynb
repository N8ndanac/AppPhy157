Define a function `gen_sinusoid` to generate a sinusoid with input arguments frequency `freq` and pixel size `px`, and output the sinusoid `w`, the intensity `intensity`, and the coordinates `x, y`



```python
def gen_sinusoid(freq, px):
  phase = 2 * np.pi / (px/freq) #Calculate the phase of the sinusoid

  #Define the coordinates
  t = np.arange(0, px, 1)
  x, y = np.meshgrid(t, t)

  w = np.sin(phase * x) #Generate a sinusoid in the horizontal (x) direction
  intensity = (1 + w)/2 #Calculate the intensity

  return w, intensity, x, y
```

```python
#Generate sinusoids with frequencies 1 hz, 3 hz, and 5 hz
w1, i1, x1, y1 = gen_sinusoid(1, 128)
w3, i3, x3, y3 = gen_sinusoid(3, 128)
w5, i5, x5, y5 = gen_sinusoid(5, 128)

#Plot the sinusoids and their 3D projections
fig = plt.figure(figsize = (24, 12))
plt.subplot(231)
plt.imshow(w1, cmap="inferno")
plt.title("1 hz")
plt.colorbar()
plt.subplot(232)
plt.imshow(w3, cmap="inferno")
plt.title("3 hz")
plt.colorbar()
plt.subplot(233)
plt.imshow(w5, cmap="inferno")
plt.title("5 hz")
plt.colorbar()
ax1 = fig.add_subplot(234, projection="3d")
ax1.plot_surface(x1, y1, i1, cmap="inferno")
ax2 = fig.add_subplot(235, projection="3d")
ax2.plot_surface(x3, y3, i3, cmap="inferno")
ax3 = fig.add_subplot(236, projection="3d")
ax3.plot_surface(x5, y5, i5, cmap="inferno")

plt.show()
```

The sinusoids were generated using
$$\sin(\phi x) = \sin\left(\frac{2\pi}{f}x\right)$$
where $\phi$ and $f$ represent the phase and frequency, respectively, of the sinusoid.
Each sinusoid has a minimum value of $-1$ and a maximum value of $1$.

One possible application of synthetic sinusoids is in phase shift profilometry. By projecting a sinusoid on a 3D surface, the waves become distorted. If we take photos of the surface with the distorted sinusoids, we can retrieve information about the 3D profile on an object. This technique is used in the 3D reconstruction of valuable artifacts, which are often fragile and cannot be subjected to 3D scanning.

Recreate the Olympics logo as an image. It is fine if the colors overlap where the rings intersect.

Define a function `draw_circle` to draw a circle on an array with the inputs radius `r`, width `width`, center `ox, oy`, and canvas `img`

In essence, the function subtracts the circle from a given canvas

```python
size = 1000 #Define the size of the canvas

def draw_circle(r, width, oy, ox, img):
  x, y = np.ogrid[:size, :size] #Take the coordinates of each point in the canvas
  t = width/2 #Take half of the thickness

  d = np.sqrt((x - ox)**2 + (y - oy)**2) #Find the distance from each coordinate to the denoted center of the circle
  img[(((r - t) <= d) * (d <= (r + t)))] = 0 #If the distance corresponds to the radius (within the thickness of the circle), subtract the pixel

  return img
```

```python
#Define parameters for each circle
radius = 150 #radius
thickness = 18 #thickness
#(x, y) coordinates for the first circle
x_o = 170
y_o = 450
#Spacing between circles
offset = 15

#Generate the canvas
I_rgb = np.ones((size, size, 3)) #Use np.ones() to begin with a white canvas
#Split the canvas into each of the R, G, B channels
rd = I_rgb[..., 0] #Red canvas
gr = I_rgb[..., 1] #Green canvas
bl = I_rgb[..., 2] #Blue

#Use the draw_circle function to subtract the circles from each canvas
#To draw a blue circle, we must subtract a circle at the same coordinates from both the Red and Green channels, leaving only the Blue
draw_circle(radius, thickness, x_o, y_o, rd)
draw_circle(radius, thickness, x_o, y_o, gr)
#Yellow = Red + Green, so we only subtract the blue
draw_circle(radius, thickness, x_o + radius + offset, y_o + radius + offset, bl)
#To create black, we must subtract from all three channels
draw_circle(radius, thickness, x_o + 2*(radius + offset), y_o, rd)
draw_circle(radius, thickness, x_o + 2*(radius + offset), y_o, gr)
draw_circle(radius, thickness, x_o + 2*(radius + offset), y_o, bl)
#To create a green circle, we must subtract from the Red and Blue channels, leaving only the Green
draw_circle(radius, thickness, x_o + 3*(radius + offset), y_o + radius + offset, rd)
draw_circle(radius, thickness, x_o + 3*(radius + offset), y_o + radius + offset, bl)
#To create a red circle, we must subtract from the Blue and Green channels, leaving only the Red
draw_circle(radius, thickness, x_o + 4*(radius + offset), y_o, bl)
draw_circle(radius, thickness, x_o + 4*(radius + offset), y_o, gr)

# Plot the RGB image
plt.figure(figsize=(5, 5), dpi=150)
plt.imshow(I_rgb)
plt.title("Olympic Rings")
plt.show()
```

We begin with a white canvas.
To create the circles, we use color subtraction. For example, to create a red circle, we subtract blue and green from the white. To create black, we subtract all channels---red, blue, and green.
As a result, in the overlaps of circles, the region becomes black.

The opposite approach can also be applied, color addition. These concepts describe the mixing of colors.

More generally, we can also add and subtract images. Image addition may be to add filters or overlays, while subtraction can be used to remove backgrounds or noise, among a variety of other applications.

Apply contrast stretching to an astronaut image.

Show the following:

3. [10 pts.] Cumulative Distribution Function (CDF) plot showing curves for the original and contrast-stretched image

First step is to load the image-- available from the scikit-image (skimage) library-- and make a low-contrast version, using the code below:

```python
from skimage import color, data, exposure
from skimage.transform import rescale

img = color.rgb2gray(data.astronaut())*256 # original astronaut image
plt.figure(figsize=(10,5), dpi = 200)
plt.subplot(121)
plt.imshow(img, cmap = 'gray', vmax = 255)
plt.colorbar()
plt.title("Original image")

img_low_contrast = img/5 # low contrast astronaut image
plt.subplot(122)
plt.imshow(img_low_contrast, cmap = 'gray', vmax = 255)
plt.colorbar()
plt.title("Low contrast image")
```

By converting the image to grayscale, the image becomes represented by a 2D matrix with dimensions 512 x 512. Each element represents a pixel, with a value ranging from 0 to 255, where 0 is black and 255 is white.

By dividing each element by 5, blacks remain at 0, however, the brightest pixels are only 1/5th as bright, with a value of 51. As a result, the contrast of the image becomes much lower.

```python
plt.figure(figsize=(15,5), dpi = 200)
plt.subplot(131)
plt.imshow(img, cmap = 'gray', vmax = 255)
plt.colorbar()
plt.title("Original image")

img_low_contrast = img/5 # low contrast astronaut image
plt.subplot(132)
plt.imshow(img_low_contrast, cmap = 'gray', vmax = 255)
plt.colorbar()
plt.title("Low contrast image")

a = 0
b = 255
c = np.min(img_low_contrast)
d = np.max(img_low_contrast)
img_contrast_stretched = (img_low_contrast - c)*((b - a)/(d - c)) + a
plt.subplot(133)
plt.imshow(img_contrast_stretched, cmap = 'gray', vmax = 255)
plt.colorbar()
plt.title("Contrast-stretched image")
```

From the low contrast image, the pixel values only ranged from 0 to 51.

By applying contrast stretching, we use the full possible range of the pixel values. Hence, we reassign all pixels with a value of 51 to 255. It works like a lookup table.

Compared to the original image, the contrast is improved. Details such as the creases on the American flag and the astronaut's spacesuit become clearer.

Next, write your code to apply contrast stretching, calculate the PDF and CDF of the original and contrast-stretched images. Finally, display the images and PDF and CDF plots.

```python
def calculate_pdf_cdf(image):
    # Flatten the image array
    flattened = image.flatten()

    # Calculate PDF (Probability Density Function)
    pdf, _ = np.histogram(flattened, bins=256, range=[0,256], density=True)

    # Calculate CDF (Cumulative Distribution Function)
    cdf = np.cumsum(pdf)

    return pdf, cdf

# Calculate PDF and CDF for original and stretched images
pdf_original, cdf_original = calculate_pdf_cdf(img_low_contrast)
pdf_stretched, cdf_stretched = calculate_pdf_cdf(img_contrast_stretched)

# Display original and stretched images
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.imshow(img_low_contrast, cmap='gray',vmax = 255)
plt.title('Low Contrast Image Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(img_contrast_stretched, cmap='gray')
plt.title('Contrast Stretched Image')
plt.axis('off')

# Plot PDF for original and stretched images
plt.subplot(2, 2, 3)
plt.plot(pdf_original, color='r', label='Low Contrast Image')
plt.plot(pdf_stretched, color='g', label='Contrast Stretched Image')
plt.title('PDF')
plt.xlabel('Gray Value')
plt.ylabel('Probability Density')
plt.fill_between(range(len(pdf_original)), pdf_original, color='r', alpha=1)
plt.fill_between(range(len(pdf_stretched)), pdf_stretched, color='g', alpha=1)
plt.legend()

# Plot CDF for original and stretched images
plt.subplot(2, 2, 4)
plt.plot(cdf_original, color='r', label='Low Contrast Image')
plt.plot(cdf_stretched, color='g', label='Contrast Stretched Image')
plt.title('CDF')
plt.xlabel('Gray Value')
plt.ylabel('Cumulative Probability')
plt.legend()
```

The probability density function is a histogram showing the distribution of pixel values. There is a peak at 0, since the image contains both dark details and shadows. For the low contrast image, there are no pixels beyond 51 gray value. On the other hand, the contrast stretched image has a much more flat distribution, ranging all the way until 255.

For the cumulative density function, the low contrast image has a very sharp increase in the interval from 0 to 51. At gray value 51, all pixels are already accounted for. On the other hand, the contrast-stretched image has an approximately linear CDF, from 0 to 255.

Physically, the compressed PDF, which zeros out early, and the compressed CDF, which maxes out early, represent low contrast. A more even distribution notes improved contrast, and more details in terms of the shadows and the highlights of the image.
