#### PART 1 - Network generation and characteristics *(30 points)*

3. Create the degree distribution plot and briefly discuss what it tells us about this social network. *(10 points)*

```python
# Import the necessary packages
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

# Mount Google Drive
from google.colab import drive
drive.mount("/content/drive")

file_path = "/content/drive/My Drive/AP157/"

# Check NetworkX version
nx.__version__
```

```python
# Read the adjlist file to create the network
G = nx.read_adjlist(file_path + 'network.adjlist', delimiter = " ")

# Visualizing
nx.draw(G, node_size=2000, node_color='mediumseagreen', with_labels=True)
```

```python
# Computing the diameter and density using NetworkX
diameter = nx.diameter(G)
density = nx.density(G)

# Get the degree distribution
degrees = [val for (node, val) in G.degree()]

# Take the average of all degrees
average_degree = np.mean(degrees)

# Output
print("Diamater: " + str(diameter))
print("Density: " + str(density))
print("Average Degree: " + str(average_degree))
```

```python
# Define the bins
degree_counts = {}

# Count each bin
for degree in degrees:
  if degree in degree_counts:
    degree_counts[degree] += 1
  else:
    degree_counts[degree] = 1

# Extract unique degrees and their counts
unique_degrees = list(degree_counts.keys())
counts = list(degree_counts.values())

# Visualizing
plt.bar(unique_degrees, counts, color="mediumseagreen")  # Unique degrees on x-axis, counts on y-axis
plt.xlabel("Degree")
plt.ylabel("No. of Nodes")
plt.title("Degree Distribution of the Social Network")
plt.show()
```

#### PART 2 - Centrality Measures *(30 points)*

Generate a table of centrality measures per node. Columns are:
- Name
- Degree Centrality
- Betweenness Centrality
- Closeness Centrality

Briefly discuss who are the most "central" people in the social network and why.

```python
import pandas as pd

nodes_arr = list(G.nodes)

df = pd.DataFrame()
df["Name"] = nodes_arr
df["Degree Centrality"] = df["Name"].map(nx.degree_centrality(G))
df["Betweenness Centrality"] = df["Name"].map(nx.betweenness_centrality(G))
df["Closeness Centrality"] = df["Name"].map(nx.closeness_centrality(G))

df
```

Jill has the highest degree centrality, which means she is the node with the most edges. In this case, Jill is connected to six other nodes.

Liz has the highest betweenness centrality, which means she is often included in the shortest path between two other members of the social network. *Siya 'yung dinadaanan ng lahat ng tsismis.*

Shane and Emma have the highest closeness centrality, which means they are in best position to share and receive information. Information from either the Liz-Allen-Lisa group or the larger close-knit group of Bob-Mike-John-Jill-Leah are both connected to Shane and Emma. *Sila 'yung mga social butterfly na kasama sa lahat ng friend group.*

#### PART 3 - Identify communities *(40 points)*

1. Identify communities in the social network using *(30 points)*:
- Louvain community detection algorithm
- Any other community detection algorithm of your choice

2. Calculate the modularity of the partitions from the two community detection algorithms. *(10 points)*

Briefly describe the algorithm you used and compare the results from the two algorithms.

```python
# Find the Louvain communities using NetworkX
Louvain_communities = nx.community.louvain_communities(G, seed=52324)

# Color each node depending on their community
color_map = []

for node in G:
  if node in Louvain_communities[0]:
    color_map.append("mediumseagreen")
  if node in Louvain_communities[1]:
    color_map.append("darkgoldenrod")
  if node in Louvain_communities[2]:
    color_map.append("firebrick")

# Visualizing
nx.draw(G, node_size=2000, node_color=color_map, with_labels=True)
plt.show()

# Computing the modularity
L_modularity = nx.community.modularity(G, Louvain_communities)
print("The modularity of the Louvain communities partition is " + str(L_modularity))
```

Three communities were identified.

```python
# Find the K-clique communities using NetworkX
KC_communities = nx.community.k_clique_communities(G, 4)

cliques = [frozenset(c) for c in KC_communities if len(c) >= 4]
loners = [] # Each node that does not belong to a community; we can describe them as belonging to a community consisiting of only the node itself

# Defining a color for each community
color_map = []

for node in G:
  if node in cliques[0] and node in cliques[1]: # If the node is included in two communities (red and yellow), color it orange :D
    color_map.append("darkorange")
  elif node in cliques[0]:
    color_map.append("gold")
  elif node in cliques[1]:
    color_map.append("firebrick")
  else:
    loners.append(node)

print("There are " + str(len(cliques)) + " communities and " + str(len(loners)) + " loners")
```

```python
# Add a unique color for each "loner" community
color_map.extend(["mediumseagreen", "palevioletred", "cornflowerblue"])

# Visualizing
nx.draw(G, node_size=2000, node_color=color_map, with_labels=True)
plt.show()
```

The network is partitioned into five different communities.

Jill is a simultaneously a member of two communities.

Liz, Allen, and Lisa do not belong to any community, and each can be considered to belong to a community consisting of only the node itself.

```python
# Jill overlaps as a member of two communities, hence we consider both possible partitions
kc1 = [["Leah", "John", "Shane"], ["Jill", "Emma", "Bob", "Mike"], ["Liz"], ["Allen"], ["Lisa"]]
kc2 = [["Leah", "John", "Shane", "Jill"], ["Emma", "Bob", "Mike"], ["Liz"], ["Allen"], ["Lisa"]]

# Computing the modularity
KC_modularity = nx.community.modularity(G, kc1) + nx.community.modularity(G, kc2)
print("The modularity of the K-clique communities partition is " + str(KC_modularity))
```

The modularity of the two partitions is almost the same, albeit the K-clique communities partition exhibiting a slightly larger value

A higher modularity denotes a better community structure.

For the Louvian communities, it is not really declared why Jill is a member of the Bob-Mike-Emma network, rather than John-Leah-Shane, when Jill is connected to all three members of both. It suggests the edges are heavier in the former network, although this is not really defined in the network, and the seeding of the code determines the result.

On the other hand, in the K-clique communities, Jill is a member of both networks. This makes sense since she is the overlap between the two groups.

The Louvian communities also consider Liz-Allen-Lisa to be one community, whereas the K-clique communities don't consider any of them to belong to any community. For a social network, this can make more sense, since Allen can be friends with both Liz and Lisa, but since Liz and Lisa aren't friends with each other, we can't really consider the three to form a friend group.
